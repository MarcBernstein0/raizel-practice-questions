### In every Olympics, which US athlete has won the most total number of medals? Include the athletes discipline.

```python
US_df  = oo[oo.NOC == 'USA']
US_df_size = US_df.groupby(['Edition', 'Athlete', 'Medal']).size()
US_df_update = US_df_size.unstack('Medal', fill_value=0)
```

```python
US_df_update['Total'] = US_df_update['Gold'] + US_df_update['Silver'] + US_df_update['Bronze']
```

```python
US_df_update.reset_index(inplace=True)
```

```python
US_df_update
```

```python
# for year,group in US_df_update.groupby('Edition'):
#     print(year)
#     print(group)
```

```python
olympics = [group.sort_values('Total',ascending=False)[:1] for year,group in US_df_update.groupby('Edition')]
```

```python
olympic_df = pd.DataFrame()
for x in olympics:
    olympic_df = olympic_df.append(x)
olympic_df.head()
```
