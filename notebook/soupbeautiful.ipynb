```python
from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get('https://en.wikipedia.org/wiki/The_Bachelor_(American_TV_series)')
# page.text
```

```python
page.status_code
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
print(soup.prettify())
```

```python
lst_links = []
for a in soup.find_all('a', href=True):
    if "/wiki/" in a['href']:
        lst_links.append(a['href'])
lst_links
```
