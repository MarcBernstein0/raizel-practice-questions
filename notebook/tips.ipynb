```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('seaborn')
```

```python
df = sns.load_dataset('iris')
```

```python
df
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
df.dropna(subset=['sepal_width'], axis =0)
```

```python
df['sepal_length_10']=df['sepal_length'] * 10
```

```python
df.groupby(['species']).count()
```

```python
df['species'].value_counts()
```

```python
df.groupby(['species'])['petal_length'].value_counts()
```

```python
petal_length_by_species = df.groupby(['species'])['petal_length'].mean()
petal_length_by_species
```

```python
x = petal_length_by_species.index
y = petal_length_by_species
figure, ax = plt.subplots(figsize = (10,6))
ax.bar(x, y, color = ['deepskyblue', 'dodgerblue', 'blue'])
ax.set_title('avg petal length', fontsize = 20)
ax.set_xlabel('species')
ax.set_xticks([0,1,2])
ax.set_xticklabels(['flower1','flower2','flower3'])
```

```python
people = ['Biden', 'Trump', 'Kanye', 'Kris']
height = [67, 62, 64, 59]
plt.bar(people,height, color = ['blue', 'orangered', 'purple', 'silver'])
```

```python
figure, ax = plt.subplots()

ax.set_title('Sepal Length vs Sepal Width')
ax.set_xlabel('Sepal Length')
ax.set_ylabel('Sepal Width')

x = df.sepal_length
y = df.sepal_width

ax.scatter(x,y)
```
